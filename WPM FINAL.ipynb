```python
df = df.drop(['SEASON', 'SAMPLE_DATE','FARMER_ENGLISH_NAME','VILLAGE_NAME_ENGLISH','SURVEY_NO','Unnamed: 7','SAMPLE_NO'], axis=1)
print(df)
```

```python

df["wpm-PH"]=df.apply(lambda row: row.PH ** 0.02449432, axis=1)  

df["wpm-EC"]=df.apply(lambda row: row.EC ** 0.02276542, axis=1)  

df["wpm-Organinc Carbon"]=df.apply(lambda row: row['Organinc Carbon'] ** 0.18277409, axis=1)  

def power(x):
    return int(x) ** 0.02449432
df["wpm-phosphorous"]=df.apply(lambda row: row.PHOSPHOROUS ** 0.1549298, axis=1)  

df["wpm-POTASH"]=df.apply(lambda row: row.POTASH ** 0.13362141, axis=1)  

df["wpm-CC"]=df.apply(lambda row: row.CALCIUM_CARBONATE ** 0.12330385, axis=1)  

df["wpm-Ferrous"]=df.apply(lambda row: row.FERROUS **  0.07798146, axis=1)  

df["wpm-Managese"]=df.apply(lambda row: row.MANAGESE **  0.08390506, axis=1)  

df["wpm-Zink"]=df.apply(lambda row: row.ZINK **  0.09771749, axis=1)  

df["wpm-Copper"]=df.apply(lambda row: row.COPPER **  0.09850709, axis=1)  

df["Results"]=df.apply(lambda row: row['PH'] * row['EC']* row['Organinc Carbon'] * row['PHOSPHOROUS']* row['POTASH']* row['COPPER'] * row['ZINK'] * row['MANAGESE'], axis=1)  
df
```

```python
df['Quality'] = pd.cut(df.Results,bins=[0,0.33,0.38,0.45,0.69],labels=['Poor','Medium','Good','Excellent'])
# print(df)

test = np.where(df['Results']>0.7)
print (test)
df['Quality'].value_counts().plot(kind='barh')
plt.show()
```
